### Идеи по Проекту

### Обработка Признаков
Так-как мы имеем дело с категориальными и числовыми признаками, то было бы неплохо обсудить **как их обрабатывать:**
- Заполнение пропущенных значений
- Кодирование (категориальные признаки)
- Масштабирование (числовые признаки)

Для кодирования категориальных признаков можно использовать
- `One-Hot Encoding` (создаем n новых признаков по m уникальным градациям категориального признака. Все компоненты таких векторов равны нулю, кроме случаев где категория m совпаадет с признаком n);
- `Dummy Variables` (аналогичен One-Hot Encoding, только используется n-1 новых признаков, чтобы избежать так Dummy Variables Trap (или по-русски проблемы мультиколлинеарности признаков для лучшей оптимизации и интерпретации модели);
- `Label Encoder` (но он вводит порядок на категории. Мы не знаем являются ли категории порядковыми, поэтому использование данного метода может быть не совсем корректно)

Для масштабирования числовых признаков:
- `StandardScaler()`
- `RobustScaler()`
- `MinMaxSclaer()`

А пропуски заполнить, используя:
- `SimpleImputer()` c различными вариантами (среднее, мода, самое частое значение)
- `KNN()`

### Кросс-Валидация
Так-как выборка несбалансированная, то будем использовать `StratifiedKFoldCrossValidation` с чисом фолдов 5. Для более объективной оценки, лучше использовать кросс-валидацию в связке с `Pipeline`, чтобы предотвратить так называемый `data leakage` и слегка оптимистичные результаты. 

### Метрики
Для оценки качества классификации в качестве основной метрики будем использовать `ROC-AUC` и ряд вспомогательных: `F1-Score`, `Precision` и `Recall` чтобы отслеживать число `False Positive` и `False Negative`
