```python
X_train = pd.read_csv("train_dataset.csv").drop(columns=['labels'])
y_train = pd.read_csv("train_dataset.csv")['labels']


cat_features = X_train.columns[190:230].values
num_features = X_train.columns[:190].values

null_percent = ((X_train.isnull().sum()/X_train.shape[0]) )
low_nulls = null_percent[null_percent<0.6]
low_nulls_features = low_nulls.reset_index()['index'].values

low_nulls_cat_features = list(set(low_nulls_features).intersection(set(cat_features)))
low_nulls_num_features = list(set(low_nulls_features).intersection(set(num_features)))

cat_features_2_15 = [category for category in low_nulls_cat_features if 2 < X_train[category].unique().size < 15]

cat_df = X_train[cat_features_2_15].fillna('other')
num_df = X_train[low_nulls_num_features].fillna(0)
low_nulls_df = pd.concat([cat_df,num_df],axis=1)

categorical_transformer = OrdinalEncoder(handle_unknown='use_encoded_value',unknown_value=-1)

preprocessor = ColumnTransformer(
    transformers=[
        ('num','passthrough',low_nulls_num_features),
        ('cat', categorical_transformer, cat_features_2_15)
    ]
)


steps = [
    ('prep',preprocessor),
    ('pipeline', imblearn.pipeline.Pipeline([
        ('undersampler',RandomUnderSampler(sampling_strategy=0.5)),
        ('clf',RandomForestClassifier(max_depth=10,n_estimators=200))
    ])),
    
]


process = sklearn.pipeline.Pipeline(steps=steps)
process.fit(low_nulls_df,y_train)


cross_val_score(process,
                low_nulls_df,
                labels,
                scoring='roc_auc',
                n_jobs=-1,
                error_score='raise'
               ).mean()
```

## Построение экономической модели

### 6.1

Оценка потенциального экономического эффекта от внедрения полученного решения

Постройте простую экономическую модель для оценки эффекта от внедрения полученного решения на практике. 

Например, введите следующие параметры:

- сколько денег в среднем приносит один пользователь в месяц;
- сколько денег в среднем вы будете вкладывать в удержание одного пользователя;
- с какой вероятностью пользователь примет ваше предложение;
- сколько пользователей (например, топ 1% или топ 25% согласно ранжированию по вашей модели) будет участвовать в кампании.

Введите еще несколько параметров и на их основе постройте простую экономическую модель: формулу по которой в зависимости от значения исходных параметров вы сможете рассчитать экономический эффект от проведения кампании (= сколько денег вы получили (или наоборот потеряли) от проведения кампании по удержанию) с использование вашей модели.  





$$
P = PPM*(TP*PA + FP-FN) - CFO*(TP+FP)
$$
PPM - сколько денег в среднем приносит один пользователь в месяц;  
CFO - сколько денег в среднем вы будете вкладывать в удержание одного пользователя;  
PA - с какой вероятностью пользователь примет ваше предложение;  
NU - сколько пользователей (например, топ 1% или топ 25% согласно ранжированию по вашей модели) будет участвовать в кампании.  
TP - сколько пользователей действиетльно относиться к классу отток из выборки NP  
FP - сколько пользователей отнесли к классу отток по ошибке из выборки NP  
FN - сколько пользователей отнесли к классу не отток по ошибке из выборки NP  



### 6.2

Исходя из введенных параметров проанализируйте, какой топ пользователей, склонных к оттоку согласно модели, оптимально выбрать для проведения кампании по удержанию для таким образом, чтобы экономический эффект был наибольшим? 

```python


def calc_formula62(NU,PPM,CFO,PA):    
    u = users.head(int(users.shape[0] * NU))

    TP = ((u['labels']==1) & (u['pred_label']==1)).sum()
    FP = ((u['labels']==-1) & (u['pred_label']==1)).sum()
    FN = ((u['labels']==1) & (u['pred_label']==-1)).sum()
    
    return  PPM*(TP*PA + FP-FN) - CFO*(TP+FP)
```

```python
PPM = 800
CFO = 200
PA = 0.9

percent = []
result = []

for i in range(1,25):
    r = calc_formula62(i*0.01,PPM,CFO,PA)
    percent.append(i)
    result.append(r)
```

```python
plt.plot(percent,result)
plt.xlabel('Процент')
plt.ylabel('Эффект')
```

```python
print(f"Максимальный экономичекий эффект достигается при {percent[np.argmax(result)]}%")
print("Максимальный эффект",np.max(result))
```

### 6.3

Попробуйте усложнить экономическую модель. Добавьте еще несколько параметров и поиграйте с ними (например, измените стоимость удержания и вероятность, с которой пользователь принимает предложение), проанализируйте как меняется оптимальный размер топа? 

Добавим еще несколько параметров

CFOA - затраты, если пользователь принял наше предложение  
CFOD - затраты, если пользователь отказался от предложения  

Обновленная формула

$$
P = PPM*(TP*PA + FP-FN) - PA*CFOA*(TP+FP) - (1-PA)*CFOD*(TP+FP)
$$

```python


def calc_formula63(NU,CFOA,CFOD,PA,PPM):    
    u = users.head(int(users.shape[0] * NU))

    TP = ((u['labels']==1) & (u['pred_label']==1)).sum()
    TN = ((u['labels']==-1) & (u['pred_label']==-1)).sum()
    FP = ((u['labels']==-1) & (u['pred_label']==1)).sum()
    FN = ((u['labels']==1) & (u['pred_label']==-1)).sum()
    
    return  PPM*(TP*PA + FP-FN) - PA*CFOA*(TP+FP) - (1-PA)*CFOD*(TP+FP)
```

```python
CFOA_values = np.linspace(20,200,25)
CFOD_values = np.linspace(10,100,25)
PA_values = np.linspace(0.1,0.9,25)
PPM = 800
result = []

p = 0.06

for CFOA,CFOD,PA in zip(CFOA_values,CFOD_values,PA_values):
    
    r = calc_formula63(p,CFOA,CFOD,PA,PPM)
    result.append(r)
```

```python
plt.plot(CFOA_values,result)
plt.xlabel('Затраты')
plt.ylabel('Эффект')
```

```python
print(f"Максимальный экономичекий эффект достигается при затратах {CFOA_values[np.argmax(result)]}")
print(f"Максимальный эффект {np.max(result):.2f}",)
print(f"Потери в случае отказа от предложения  {CFOD_values[np.argmax(result)]}")
print(f"Пользователь согласится на предложение с вероятностью  {PA_values[np.argmax(result)]:.2f}%")
```

Можно сделать вывод, что оптимальной суммой для удержания пользователя является 8.4% от прибыли которую он приносит в месяц

### 6.5

 Оцените изменение экономического эффекта от проведения кампании по удержанию при увеличении качества модели на 1%? На 3%? При ответе на вопрос укажите, по какой метрике вы оцениваете качество.

В данном случае качество модели определяется количеством ошибок

FN - модель пропустила пользователя собравшегося уходит. Действия по удержанию не проведены. (большие потери)

FP - проводим действия по удержанию для пользователя ошибочно отнесенного к "уходящим". (небольшие потери)

Соответственно, если качество модели увеличивается на 1-3%, то количество таких ошибок уменьшается.

```python
u = users

FP = ((u['labels']==-1) & (u['pred_label']==1)).sum()
FN = ((u['labels']==1) & (u['pred_label']==-1)).sum()
PA = 0.5

print('Всего на выборке FN:', FN)
print('Всего на выборке FP:', FP)
```

Если использовать простейшую модель для оценки, то каждая FN это убыток равный упущенной прибыли (PPM * PA) из-за бездействия.

Каждая FP - убыток из-за затрат на проведение кампании с пользователем, который и так не собирался уходить (CFO).

```python
print ('Для улучшения качества модели на 1%:')
print ('Уменьшение количества FN:', int(FN * 0.01))
print ('Уменьшение количества FP:', int(FP * 0.01))
print ('Уменьшение убытков из-за улучшения качества модели на 1%:', int(FN * 0.01) * PPM * PA + int(FP * 0.01) * CFO)
```

```python
print ('Для улучшения качества модели на 3%:')
print ('Уменьшение количества FN:', int(FN * 0.03))
print ('Уменьшение количества FP:', int(FP * 0.03))
print ('Уменьшение убытков из-за улучшения качества модели на 3%:', int(FN * 0.03) * PPM * PA + int(FP * 0.03) * CFO)
```
