### Необходимые Функции

```python
# Удаляем признаки, состоящие только из NaN 
def drop_nan_columns(df):
    """
    Drops columns that have only NaN values
    
    """
    feat_to_drop = [feature for feature in df.columns if df[feature].isna().all()]

    print(f'Dropped {len(feat_to_drop)} NaN Columns')
    df.drop(columns=feat_to_drop, axis=1, inplace=True)  
    
# Возвращает названия столбцов для бинарных и категориальных признаков 
def return_bin_cat_columns(df):
    bin_columns = [feature for feature in df.columns if df[feature].value_counts().shape[0] == 2]
    cat_columns = df.columns.difference(df[bin_columns].columns).to_list()
    
    return bin_columns, cat_columns

# Гистограмма пропусков
def plot_nan_ration(df, figsize=(10,5), title='Numerical Features'):
    fig = plt.figure(figsize=figsize)
    plt.hist(df.isna().sum()/df.shape[0])
    plt.xlabel('NaN Fraction')
    plt.ylabel('Number of Features')
    plt.title(title)
    plt.grid();    
```

**1)** Начнем с простого. Давайте оценим как много объектов действительно нужно для построения качественной модели. Для обучения доступна достаточно большая выборка и может так оказаться, что начиная с некоторого момента рост размера обучающей выборки перестает влиять на качество модели. **Постройте кривые обучения**, обучая модель на выборках разного размера начиная с небольшого количество объектов в обучающей выборке и постепенно наращивая её размер с некоторым шагом. Обратите внимание на `sklearn.model_selection.learning_curve`

Будем по-прежнему использовать `Pipeline`. Бинарные и категориальные признаки объединим в один блок, и будем применять к ним одни и те же трансформации. Это объясняется тем, что для некоторых бинарных признаков также имеются пропуски, и если заменить их на `unknown`, то они уже перестают быть бинарными и становятся категориальными, чтобы не усложнять, мы их объеденим.

На данном этапе импьютеры, масштабирование и энкодеры категориальных признаков выбраны абсолютно случайно, позже выясним, что является лучшим вариантом

```python
# Далее очень часто придется собирать различные Pipeline, поэтому создадим функцию для сборки определенного Pipeline
def make_pipe(cat_bin_columns,
              num_columns,
              model=None,
              cat_bin_imputer=SimpleImputer(strategy='constant', fill_value='unknown'),
              cat_bin_encoder=OneHotEncoder(sparse=True, handle_unknown='ignore'),
              num_imputer=SimpleImputer(strategy='mean'),
              num_scaler=StandardScaler()):
    
    # Categorical/Binary Features Processing
    cat_bin_pipeline = Pipeline([
        ('cat_bin_imputing', cat_bin_imputer),
        ('cat_bin_encoding', cat_bin_encoder)
    ])
    
    # Numerical Features Processing
    num_pipeline = Pipeline([
        ('data_imputing', num_imputer),
        ('data_scaling', num_scaler)
    ])
    
    # Main Transformations
    transformations = [
        ('cat_bin_transformations', cat_bin_pipeline, cat_bin_columns),
        ('num_transformations', num_pipeline, num_columns)
    ]
    
    feature_transformations = ColumnTransformer(transformers=transformations, n_jobs=-1)
    
    main_pipeline = Pipeline([
        ('feature_transformations', feature_transformations)
    ])
    
    if model is not None:
        main_pipeline.steps.insert(1, ('model', model))
        return main_pipeline
    else:
        return main_pipeline
```

```python
# Создаем Pipeline
model_pipeline = make_pipe(cat_bin_columns=cat_features.columns,
                           num_columns=num_features.columns,
                           model=GradientBoostingClassifier(random_state=SEED),
                           cat_bin_imputer=SimpleImputer(strategy='constant', fill_value='unknown'),
                           cat_bin_encoder=OneHotEncoder(sparse=True, handle_unknown='ignore'),
                           num_imputer=SimpleImputer(strategy='mean'),
                           num_scaler=StandardScaler())
```

```python
# Function for plotting Learning Curves
def plot_with_std(n_samples, data, **kwargs):
    mu, std = data.mean(1), data.std(1)
    lines = plt.plot(n_samples, mu, '-', **kwargs)
    plt.fill_between(n_samples, mu - std, mu + std, edgecolor='none', facecolor=lines[0].get_color(), alpha=0.2)

def plot_learning_curve(model, x_train, y_train, cv_type, scorer, train_data_split=20, shuffle=True):
    train_sizes = np.linspace(0.05, 1, train_data_split)
    n_train, train_curve, val_curve = learning_curve(estimator=model, X=x_train, y=y_train,
                                                     cv=cv_type, scoring=scorer, train_sizes=train_sizes,
                                                     n_jobs=-1, shuffle=shuffle, random_state=SEED)
    plt.figure(figsize=(20,8))
    plot_with_std(n_train, train_curve, label='Training Scores')
    plot_with_std(n_train, val_curve, label='Validation Scores')
    plt.xlabel('Training Set Size')
    plt.ylabel(scorer)
    plt.legend()
    plt.grid()
```

Веса при соответствующих объектах задаются при помощи параметра `sample_weight`, принимающий массив из весов. 

Так-как мы используем модуль `Pipeline`, то согласно документации необходимо использовать `step_name__sample_weight`, что в нашем случае `model__sample_weight`. В зависимости от того кокой подход используется (Кросс-валидация или `GridSearchCV`), необходимо по-разному передавать параметры

**Кросс-валидация**

Необходимо передать в параметр `fit_params` словарь вида: `{'model__sample_weight': weights_array}` 

**GridSearchCV**

Необходимо передать в метод `fit` объекта `GridSearchCV` параметр `model__sample_weight=weights_array`

**Мы же будем использовать кросс-валидацию**, чтобы отследить как изменяются основные метрики при изменении весов

```python
# Определим функцию в которой можно быстро смотреть как меняется качесвто на конкретном множестве весов
def show_model_cvs_class_weights(model, x_train, y_train, cv_type, metrics, weights=np.linspace(0.05, 0.95, 20)):
    res_df = pd.DataFrame()
    
    for weight in log_progress(weights):
        cv_results = cross_validate(model,
                                    X=x_train,
                                    y=y_train, 
                                    cv=cv_type,
                                    scoring=metrics,
                                    error_score='raise',
                                    n_jobs=-1,
                                    fit_params={'model__sample_weight': y_train.apply(lambda x: weight if x == 1.0 else 1 - weight)})
           
        cv_results['Min_Class_Weight'] = weight 
        res_df = res_df.append(pd.DataFrame(pd.DataFrame(cv_results).mean()).iloc[2:, :].T)
    
    return res_df
```

Объект, принадлежащий целевому классу `+1` получает значение `weight`, а объект класса `-1` значение `1-weight`

```python
%%time
weights_cv_results = show_model_cvs_class_weights(
    model=model_pipeline,
    x_train=x_train,
    y_train=y_train,
    cv_type=StratifiedKFold(shuffle=True, random_state=SEED),
    metrics=['precision', 'recall', 'f1', 'roc_auc'],
    weights=np.linspace(0.05, 0.95, 20)
)

weights_cv_results = weights_cv_results.reset_index(drop=True)
```

Изменение основных метрик в зависимостиот весов объектов

### Random Undersampling
Процедуру random oversampling/undersampling можно провести при помощи:
- Модуля `imblearn`
- Самостоятельно, используя метод `pd.sample`

Число элементов мажоритарного класса `К` которые необходимо удалить (в нашем случае мы будем просто набирать новую случайную выборку из объектов мажориторного класса, используя метод `pd.sample`) будем подбирать на кросс-валидации.

Перед тем как использовать различные варианты сымплинга, **хотелось бы взглянуть как выглядят наши классы в пространстве признаков.**
- Хорошо ли они разлчичимы? 
- Как классы пересекаются между собой?

Для этого используем `PCA` с понижением исходного многомерного пространства признаков в двумерное. Для использования PCA необходимо:
- Избавиться от пропусков
- Закодировать категориальные и бинарные признаки 
- Отмасштабировать числовые признаки

Использовать `Pipeline` здесь уже не получится, поэтому проделаем данную работу "руками". 

Создадим функцию, которая будет слегка громоздкой, но полезной.

```python
from category_encoders.cat_boost import CatBoostEncoder
from sklearn.decomposition import PCA

# Функция для визуализации многомерной выборки
def plot_multidim_data(x_train,
                       y_train, 
                       cat_bin_imputer=SimpleImputer(strategy='most_frequent'),
                       cat_bin_encoder=CatBoostEncoder(random_state=SEED),
                       num_imputer=SimpleImputer(strategy='mean'),
                       num_scaler=StandardScaler(),
                       dim_red_method=PCA(n_components=2),
                       sampling_method=None,
                       fig_size=(12,6),
                       title='2D Feature Space (PCA)'):
    """
    x_train - matrix of features that includes nominal features (categorical and binary features) and continious features
    y_train - target feature
    
    x_train is then splitted according features types. Each feature is then processed separately
    
    Continuous Features must have int64 or float64 data type;
    Nominal Features must have data type Object;
    
    """
    
    # Features Selection
    num_features = x_train.select_dtypes(include=['float64', 'int64'])
    cat_bin_features = x_train.select_dtypes(include=['object'])
    
    # Categorical/Binary Features Processing 
    cat_bin_features = cat_bin_imputer.fit_transform(cat_bin_features)
    cat_bin_features = cat_bin_encoder.fit_transform(cat_bin_features, y_train)
    
    # Numerical Features Processing
    num_features = num_imputer.fit_transform(num_features)
    num_features = num_scaler.fit_transform(num_features)

    # Final Matrix of Features
    final_matrix = np.hstack((num_features, cat_bin_features))
    
    target_name = y_train.name
    
    # If there is a sampling method (over/down)
    if sampling_method:
        final_matrix, y_train = sampling_method.fit_resample(final_matrix, y_train.values)
        
    final_matrix_2d = pd.DataFrame(dim_red_method.fit_transform(final_matrix))
    final_matrix_2d[target_name] = y_train
    
    print('Number of Samples (Negative Class): ', final_matrix_2d[final_matrix_2d[target_name] == np.unique(y_train).min()].shape[0])
    print('Number of Samples (Positive Class): ', final_matrix_2d[final_matrix_2d[target_name] == np.unique(y_train).max()].shape[0])
    
    # Plotting
    plt.figure(figsize=fig_size)
    plt.scatter(final_matrix_2d[final_matrix_2d[target_name] == np.unique(y_train)[0]][0], 
                final_matrix_2d[final_matrix_2d[target_name] == np.unique(y_train)[0]][1],
                label=np.unique(y_train)[0])
    
    plt.scatter(final_matrix_2d[final_matrix_2d[target_name] == np.unique(y_train)[1]][0], 
                final_matrix_2d[final_matrix_2d[target_name] == np.unique(y_train)[1]][1],
                label=np.unique(y_train)[1])
    plt.xlabel('First Component')
    plt.ylabel('Second Component')
    plt.title(title)
    plt.grid()
    plt.legend()
```

```python
%%time
plot_multidim_data(x_train=x_train,
                   y_train=y_train, 
                   cat_bin_imputer=SimpleImputer(strategy='constant', fill_value='unknown'),
                   cat_bin_encoder=CatBoostEncoder(random_state=SEED),
                   num_imputer=SimpleImputer(strategy='mean'),
                   num_scaler=StandardScaler(),
                   dim_red_method=PCA(n_components=2, random_state=SEED),
                   fig_size=(25,15), 
                   title='2D Feature Space (PCA)')
```

Видно, что классы очень сильно перемешаны между собой, также имеются выбросы для клиентов склонных и не склонных к оттоку.

Данная картина может дать неплохие представления о том какие методы under/oversamplig можно использовать, а какие нет...ведь все зависит от данных, и некоторые методы могут только навредить.

```python
# Определяем функцию over/under sampling 
def show_model_cvs_random_sampling(model, x_train, y_train, cv_type, metrics,
                                   sample_iters=20, sampling_type='under',
                                   target_labels=[-1, 1], random_state=SEED):
    res_df = pd.DataFrame()
    
    # Определяем число объектов мажоритарного и минорного классов
    maj_class_size, min_class_size = y_train.value_counts()
    
    # Поделим данный интервал на необходимое количество значений
    n_samples = np.linspace(min_class_size, maj_class_size, sample_iters)
    n_samples = np.floor(n_samples).astype('int') # Получим целочисленные значения
    
    # Разделяем выборки на классы
    x_train[y_train.name] = y_train
    
    x_train_pos_class = x_train[x_train[y_train.name] == max(target_labels)] 
    x_train_neg_class = x_train[x_train[y_train.name] == min(target_labels)]
    
    x_train.drop(y_train.name, axis=1, inplace=True) # Чтобы после завершения, таргет не присутствовал в обучении
    
    if sampling_type == 'under':
        for sample_size in log_progress(n_samples):
            # Сэмлируем необходимое число объектов мажоритарного класса
            x_train_neg_under = x_train_neg_class.sample(sample_size, random_state=random_state)

            # Cоздаем обучающую выборку
            x_train_under = pd.concat([x_train_neg_under, x_train_pos_class], axis=0)
            y_train = x_train_under[y_train.name]

            # Удаляем целевой класс из обучения
            x_train_under.drop(y_train.name, axis=1, inplace=True)

            # Кросс-валидируемся
            cv_results = cross_validate(model, X=x_train_under, y=y_train, 
                                        cv=cv_type, scoring=metrics, error_score='raise', n_jobs=-1)

            cv_results['Sample_Size'] = sample_size
            res_df = res_df.append(pd.DataFrame(pd.DataFrame(cv_results).mean()).iloc[2:, :].T)
            
        return res_df
    
    else:
        for sample_size in log_progress(n_samples):
            # Сэмлируем необходимое число объектов минорного класса
            x_train_pos_over = x_train_pos_class.sample(sample_size, replace=True, random_state=random_state)
            
            # Cоздаем обучающую выборку
            x_train_over = pd.concat([x_train_neg_class, x_train_pos_over], axis=0)
            y_train = x_train_over[y_train.name]
        
            # Удаляем целевой класс из обучения
            x_train_over.drop(y_train.name, axis=1, inplace=True)
        
            # Кросс-валидируемся
            cv_results = cross_validate(model, X=x_train_over, y=y_train, 
                                        cv=cv_type, scoring=metrics, error_score='raise', n_jobs=-1)
        
            cv_results['Sample_Size'] = sample_size
            res_df = res_df.append(pd.DataFrame(pd.DataFrame(cv_results).mean()).iloc[2:, :].T)
            
        return res_df
```

```python
%%time 
under_cv_results = show_model_cvs_random_sampling(model=model_pipeline,
                                                  x_train=x_train,
                                                  y_train=y_train,
                                                  cv_type=StratifiedKFold(shuffle=True, random_state=SEED),
                                                  metrics=['precision', 'recall', 'f1', 'roc_auc'],
                                                  sampling_type='under',
                                                  sample_iters=20)

under_cv_results = under_cv_results.reset_index(drop=True)
```

```python
# Изменение ROC-AUC
plt.figure(figsize=(12,6))
plt.plot(under_cv_results['Sample_Size'], under_cv_results['test_roc_auc'])
plt.xlabel('Majority Class Sample Size')
plt.ylabel('ROC-AUC')
plt.grid();

# Изменение F1-Score
plt.figure(figsize=(12,6))
plt.plot(under_cv_results['Sample_Size'], under_cv_results['test_f1'])
plt.xlabel('Majority Class Sample Size')
plt.ylabel('F1-Score')
plt.grid();
```

### Tomek Links

**Метод Tomek Links** является одним из алгоритмов undersampling. Данный метод позволяет удалять объекты минорного и мажоритарного класса, которые являются либо шумовыми, либо находящимися на границе классов (decision boundary). Это способствует очищению от шума и лучшей разделимости классов.

В нашем случае метод должен помочь, так-как наши классы сильно перемешаны, граница классов также сложна и имеются шумовые наблюдения 

**Важно:** если мы хотим исползовать методы сэмплинга, используя `Pipeline`, то необходимо ипортировать `Pipeline` из модуля `imblearn` 

```python
from imblearn.pipeline import Pipeline
from imblearn.under_sampling import TomekLinks

# Переопределяем Pipeline
model_pipeline = make_pipe(cat_bin_columns=cat_features.columns,
                           num_columns=num_features.columns,
                           model=GradientBoostingClassifier(random_state=SEED),
                           cat_bin_imputer=SimpleImputer(strategy='constant', fill_value='unknown'),
                           cat_bin_encoder=OneHotEncoder(sparse=True, handle_unknown='ignore'),
                           num_imputer=SimpleImputer(strategy='mean'),
                           num_scaler=StandardScaler())

# Добавим этап downsampling в Pipeline
model_pipeline.steps.insert(1, ('downsampling', TomekLinks(sampling_strategy='majority', n_jobs=-1)))
```

```python
%%time
cv_results = cross_validate(model_pipeline, X=x_train, y=y_train, 
                            cv=StratifiedKFold(shuffle=True, random_state=SEED),
                            scoring=['precision', 'recall', 'f1', 'roc_auc'], error_score='raise', n_jobs=-1)
        
pd.DataFrame(pd.DataFrame(cv_results).iloc[:,2:].mean()).T
```

### One-Sided Selection
Данный метод комбинирует такие методы как `Tomek Links` и `Condensed Neighbour (CNN)`.
- CNN удаляет большие сгустки мажоритарного класса
- Tomek Links "убирает шум" и делает границу классов более разделимой

```python
%%time

from imblearn.under_sampling import OneSidedSelection

# Параметры алгоритма котоыре переберем
n_neighbors = [3, 6, 9, 12]
n_seeds_S = [1, 3, 5, 7]

# Для хранения результатов кросс-валидации 
res_df = pd.DataFrame()

for neighbors in log_progress(n_neighbors):
    for seeds in log_progress(n_seeds_S):
        model_pipeline.steps[1] = ('downsampling', OneSidedSelection(sampling_strategy='majority',
                                                                     n_neighbors=neighbors,
                                                                     n_seeds_S=seeds, random_state=SEED, n_jobs=-1))
        
        cv_results = cross_validate(model_pipeline, X=x_train, y=y_train, 
                                    cv=StratifiedKFold(shuffle=True, random_state=SEED),
                                    scoring=['precision', 'recall', 'f1', 'roc_auc'], error_score='raise', n_jobs=-1)
        
        current_cv = pd.DataFrame(pd.DataFrame(cv_results).iloc[:,2:].mean()).T
        current_cv['n_neighbors'] = neighbors
        current_cv['n_seeds_S'] = seeds
        
        res_df = res_df.append(current_cv)
        
res_df = res_df.reset_index(drop=True)
res_df
```

Качество не особо меняется, максимум достигается на 7-ой итерации, что чуть больше чем у `TomekLinks`

### SMOTE
Данный алгоритм генерирует новые объекты минорного класса которые были бы максимально похожи на объекты минорного класса, но слегка отличались.

Минусом данного подхода является тот факт, что алгоритм вслепую увеличивает плотность примерами в области слабо представленного класса и следовательно еще сильнее перемешивает классы.

В параметр `sampling_strategy` передадим словарь с найденным соотношением классов, обеспечивающих максимальное качество. Данный метод имеет гиперпараметр `k_neighbors` который мы подбираем, используя кросс-валидацию.

```python
%%time
from imblearn.over_sampling import SMOTE

neighbours = [3, 6, 9]

smote_cv_res = pd.DataFrame()

for neighbour in neighbours:
    model_pipeline.steps[1] = ('oversampling', SMOTE(sampling_strategy={-1.0: 16921, 1.0: 9558},
                                                     k_neighbors=neighbour, random_state=SEED, n_jobs=-1))
    
    cv_results = cross_validate(model_pipeline, X=x_train, y=y_train, 
                                cv=StratifiedKFold(shuffle=True, random_state=SEED),
                                scoring=['precision', 'recall', 'f1', 'roc_auc'], error_score='raise', n_jobs=-1)
    
    cv_results = pd.DataFrame(pd.DataFrame(cv_results).iloc[:,2:].mean()).T 
    cv_results['k_neighbors'] = neighbour
    
    smote_cv_res = smote_cv_res.append(cv_results)

smote_cv_res = smote_cv_res.reset_index(drop=True)
smote_cv_res
```

Оптимальным является использование `k_neighbors=6`

При этом мы должны учитывать, что мы имеем разные типы признаков, категориальные и числовые, пропуски которых должны обрабатываться по-разному

Выберем несколько методов:
- `SimpleImputer` c параметрами `most_frequent` и `constant` (категории), `median` и `mean`  (числовые)
- `SimpleImputer` c параметрами `most_frequent` и `constant` (категории), `KNNImputer` (числовые)
- `SimpleImputer` c параметрами `most_frequent` и `constant` (категории), `SimpleImputer` c константой `-1e9` (числовые)

Кодирование для категориальных признаков пока оставим по умолчанию (`OHE`), а для числовых поэксперементируем с:
- `StandardScaler` 
- `MinMaxScaler`
- `RobustScaler`

```python
%%time

cat_bin_imputers = [
    SimpleImputer(strategy='most_frequent'),
    SimpleImputer(strategy='constant', fill_value='unknown')
]

num_imputers = [
    SimpleImputer(strategy='median'),
    SimpleImputer(strategy='mean'),
    SimpleImputer(strategy='constant', fill_value=-1e9),
    KNNImputer()
]

scalers = [
    StandardScaler(),
    MinMaxScaler(),
    RobustScaler()
]

imputers_cv_res = pd.DataFrame()

for cat_bin_imputer in log_progress(cat_bin_imputers):
    for num_imputer in log_progress(num_imputers):
        for scaler in log_progress(scalers):
            model_pipeline = make_pipe(cat_bin_columns=cat_features.columns,
                                       num_columns=num_features.columns,
                                       model=GradientBoostingClassifier(random_state=SEED),
                                       cat_bin_imputer=cat_bin_imputer,
                                       cat_bin_encoder=OneHotEncoder(sparse=True, handle_unknown='ignore'),
                                       num_imputer=num_imputer,
                                       num_scaler=scaler)
            
            cv_results = cross_validate(model_pipeline,
                                        X=x_train,
                                        y=y_train,
                                        cv=StratifiedKFold(shuffle=True, random_state=SEED),
                                        scoring=['precision', 'recall', 'f1', 'roc_auc'],
                                        error_score='raise',
                                        n_jobs=-1)

            cv_results = pd.DataFrame(pd.DataFrame(cv_results).iloc[:,2:].mean()).T

            cv_results['cat_bin_imputer'] = cat_bin_imputer
            cv_results['num_imputer'] = num_imputer
            cv_results['num_scaler'] = scaler
            
            imputers_cv_res = imputers_cv_res.append(cv_results)
        
        
imputers_cv_res = imputers_cv_res.reset_index(drop=True)
imputers_cv_res       
```

Можно заметить, что для числовых признаков лучше всего обрабатывать пропуски, используя среднее значение. Для категориальных признаков, можно использовать моду либо оформлять пропуск как отдельную категорию `unknown`.

Даже не смотря на то что заполнение пропущенных значений в категориях модой приводит к чуть лучшему качеству, я выберу заполенение неизвестной категорией (думаю так более правильно, + есть признаки где доля пропусков больше 50-60%).

Далее всегда используем следующие импутеры:
- Категориальные признаки: `SimpleImputer(strategy='constant', fill_value='unknown')`
- Числовые признаки: `SimpleImputer(strategy='mean')`
- Mасштабирование: `MinMaxScaler`

При кодировании категориальных признаков, используя `Pipeline`, не все так прозрачно. Возникает **проблема**, что при кодировании **некоторые значения становятся `inf`, это воспринимается как пропущенное значение и код перестает работать.**

Пришлось добавить замену пропусков/inf на 0 еще раз в основной функции (`final_imputing`), после этого все успешно заработало.

Так-как функцию приходится очень часто переопределять, то нужен другой, более гибкий способ конфигурирования `Pipeline`, но пока я не стал заморачиваться))

```python
# Переопределим функцию make_pipe, добавив блок final_imputing ( необходимо добавить, т.к. появляется inf, которую заменяем на 0)
def make_pipe(cat_bin_columns,
              num_columns,
              model=None,
              cat_bin_imputer=SimpleImputer(strategy='constant', fill_value='unknown'),
              cat_bin_encoder=OneHotEncoder(sparse=True, handle_unknown='ignore'),
              num_imputer=SimpleImputer(strategy='mean'),
              num_scaler=StandardScaler()):
    
    # Categorical/Binary Features Processing
    cat_bin_pipeline = Pipeline([
        ('cat_bin_imputing', cat_bin_imputer),
        ('cat_bin_encoding', cat_bin_encoder),
        ('final_imputing', SimpleImputer(strategy='constant', fill_value=0)) # Полученные inf заменяем на 0
    ])
    
    # Numerical Features Processing
    num_pipeline = Pipeline([
        ('num_imputing', num_imputer),
        ('num_scaling', num_scaler)
    ])
    
    # Main Transformations
    transformations = [
        ('cat_bin_transformations', cat_bin_pipeline, cat_bin_columns),
        ('num_transformations', num_pipeline, num_columns)
    ]
    
    feature_transformations = ColumnTransformer(transformers=transformations, n_jobs=-1)
    
    main_pipeline = Pipeline([
        ('feature_transformations', feature_transformations)
    ])
    
    if model is not None:
        main_pipeline.steps.insert(1, ('model', model))
        return main_pipeline
    else:
        return main_pipeline
```

```python
%%time

cat_encoders = [
    TargetEncoder(),
    OneHotEncoder(sparse=True, handle_unknown='ignore'),
    CatBoostEncoder(random_state=SEED),
    LeaveOneOutEncoder(random_state=SEED)
]

encoder_cv_res = pd.DataFrame()

for cat_encoder in cat_encoders:
    model_pipeline = make_pipe(cat_bin_columns=cat_features.columns,
                               num_columns=num_features.columns,
                               model=GradientBoostingClassifier(random_state=SEED),
                               cat_bin_imputer=SimpleImputer(strategy='constant', fill_value='unknown'),
                               cat_bin_encoder=cat_encoder,
                               num_imputer=SimpleImputer(strategy='mean'),
                               num_scaler=MinMaxScaler())
    
    cv_results = cross_validate(model_pipeline,
                                X=x_train,
                                y=y_train,
                                cv=StratifiedKFold(shuffle=True, random_state=SEED),
                                scoring=['precision', 'recall', 'f1', 'roc_auc'],
                                error_score='raise',
                                n_jobs=-1)
    
    cv_results = pd.DataFrame(pd.DataFrame(cv_results).iloc[:,2:].mean()).T
    cv_results['cat_encoder'] = cat_encoder
    encoder_cv_res = encoder_cv_res.append(cv_results)
    
               
encoder_cv_res = encoder_cv_res.reset_index(drop=True)
encoder_cv_res 
```

### Категориальные признаки 
Для отбора категориальных признаков воспользуемся расчетом значения V-Краммера 

```python
import scipy.stats as sts

# Определим функцию для подсчета коэффициента V-крамера (стандартная формула)
def calculate_crammer_coef(feat_1, feat_2, correction=True, return_p_value=True):
    """
    feat_1 - independent variable
    feat_2 - dependent variable
    
    if return_p_value=True returns p-value and V-Crammer (tuple) else only V-Crammer value (scalar)
    
    Returns feature name and its V-Crammer Value
    
    """
    confusion_matrix = pd.crosstab(feat_1, feat_2)
    n_observations = feat_1.shape[0]
    n_rows, n_colmns = confusion_matrix.shape
    
    # Confusion Matrix должна соответствовать ряду условий перед применением метода
    ratio = (np.sum((confusion_matrix.loc[:,-1] < 5)) + np.sum((confusion_matrix.loc[:,-1] < 5)))/confusion_matrix.size
    
    if ratio <= 0.2:
        # Если нужны p-values
        if return_p_value:
            chi2_res = sts.chi2_contingency(confusion_matrix, correction=correction)
            return feat_1.name, chi2_res[1], np.sqrt(chi2_res[0]/(n_observations*(min(n_rows, n_colmns)-1)))
        else:
            chi2_val = sts.chi2_contingency(confusion_matrix, correction=correction)[0]
            return feat_1.name, np.sqrt(chi2_val/(n_observations*(min(n_rows, n_colmns)-1)))
```

```python
# Вычисляем корреляции для категориальных признаков
crammer_corr_values = []
for feature in cat_features.columns:
    crammer_data = calculate_crammer_coef(feat_1=cat_features[feature],
                                          feat_2=y_train,
                                          correction=True,
                                          return_p_value=True)
    if crammer_data:
        crammer_corr_values.append(crammer_data)
        
# Оформим все в DataFrame
cat_feat_corrs = pd.DataFrame(crammer_corr_values, columns=['Feature', 'p-value', 'V_Crammer_Value']).sort_values(by='V_Crammer_Value',ascending=False)
cat_feat_corrs
```

```python
%%time

# Результаты кросс-валидации
feat_sel_model_cv_res = pd.DataFrame()

for estimator in estimators:
    model_pipeline.steps.insert(1, ('feature_selection', SelectFromModel(estimator)))
    
    cv_results = cross_validate(model_pipeline, X=x_train, y=y_train,
                                cv=StratifiedKFold(shuffle=True, random_state=SEED),
                                scoring=['precision', 'recall', 'f1', 'roc_auc'],
                                error_score='raise',
                                n_jobs=-1)
    
    cv_results = pd.DataFrame(pd.DataFrame(cv_results).iloc[:, 2:].mean()).T
    cv_results['selector_model'] = estimator
    
    feat_sel_model_cv_res = feat_sel_model_cv_res.append(cv_results)        
    model_pipeline.steps.pop(1)
    
             
feat_sel_model_cv_res = feat_sel_model_cv_res.reset_index(drop=True)
feat_sel_model_cv_res   
```

Теперь остались признаки с хоть какой-то изменчивостью, однако в них по-прежнему присутствуют выбросы. Найдем такие наблюдения и заменим их на NaN, а затем применим импьютацию. Посмотрим как это повлияет на качество.

```python
def remove_outliers(df_in, col = '', strategy='fill_na'):
    """
    Calculates quantiles for outliers + interquartile range. Everything above and below the borders will not be included
    in final dataframe.
    
    """
    Q1 = df_in[col].quantile(0.25)
    Q3= df_in[col].quantile(0.75)
    IRQ = Q3 - Q1
    upper_border = Q3 + IRQ*1.5
    lower_border = Q1 - IRQ*1.5
    
    if strategy == 'fill_na':
        mask = (df_in[col] >= lower_border) & (df_in[col] <= upper_border)
        df_in[col] = df_in[col].where(mask)
        
        return df_in[col]

# Отбираем выбросы и зменяем их на NaN
res_df = pd.DataFrame()

for feature in num_features_selected.columns:
    current_df = remove_outliers(num_features_selected, col=feature)
    res_df = pd.concat([res_df, current_df], axis=1)
```

### HyperOpt

```python
from hyperopt import tpe, hp, fmin, space_eval, Trials
from hyperopt.pyll import scope as ho_scope
from hyperopt.pyll.stochastic import sample as ho_sample
```

```python
model_pipeline = make_pipe(cat_bin_columns=x_train.select_dtypes(['object']).columns.to_list(),
                           num_columns=x_train.select_dtypes(['int64', 'float64']).columns.to_list(),
                           model=GradientBoostingClassifier(random_state=SEED),
                           cat_bin_imputer=SimpleImputer(strategy='constant', fill_value='unknown'),
                           cat_bin_encoder=TargetEncoder(),
                           num_imputer=SimpleImputer(strategy='mean'),
                           num_scaler=StandardScaler())
```

```python
# Так будет удобней
class ModelHyperparameters:
    
    def __init__(self, model, x, y, params_space, n_trials, cv_metric, cv_type, fit_params, opt_algo=tpe.suggest, seed=23):
        self.model = model
        self.x = x
        self.y = y
        self.params_space = params_space
        self.n_trials = n_trials
        self.cv_metric = cv_metric
        self.cv_type = cv_type
        self.fit_params = fit_params
        self.opt_algo = opt_algo
        self.seed = seed
        self.trials = Trials()
        
    def _objective(self, params):
        self.model.set_params(**params)
        
        cv_score = cross_validate(estimator=self.model,
                                  X=self.x,
                                  y=self.y,
                                  cv=self.cv_type,
                                  scoring=self.cv_metric,
                                  error_score='raise',
                                  fit_params=self.fit_params,
                                  n_jobs=-1)
        
        auc_score = pd.DataFrame(pd.DataFrame(cv_score).mean()).iloc[2:, :].T['test_score'][0]
                                                    
        return -auc_score
        
    def optimize(self):
        return fmin(fn=self._objective,
                    space=self.params_space,
                    algo=self.opt_algo,
                    max_evals=self.n_trials,
                    trials=self.trials,
                    rstate=np.random.RandomState(self.seed))
```

Оказалось, что **не все параметры стоит подбирать**, т.к. некоторые параметры являются очень хорошими по-умолчанию и их изменение приводит к снижению качества.

```python
# Пространство параметров 
params_space = {
    'model__learning_rate': hp.loguniform('learning_rate', np.log(0.001), np.log(0.5)),
    'model__n_estimators': ho_scope.int(hp.quniform('n_estimators', 50, 500, 1)),
    'model__criterion': hp.choice('criterion', ['friedman_mse', 'mse']), # don't use mae
    'model__min_samples_split': hp.loguniform('min_samples_split', np.log(0.1), np.log(1)),
    'model__max_depth':  ho_scope.int(hp.quniform('max_depth', 1, 8, 1)),
    'model__max_features': ho_scope.int(hp.quniform('max_features', 1, x_train.shape[1], 1)),
    'model__random_state': SEED
} 
```

```python
# Определяем объект model_hyperparameters
model_hyperparameters = ModelHyperparameters(model=model_pipeline, x=x_train, y=y_train,
                                             params_space=params_space, n_trials=50, cv_metric='roc_auc',
                                             fit_params=fit_params,
                                             cv_type=StratifiedKFold(shuffle=True, random_state=SEED))

# Находим наилучшее сочетание
best_params_hp = model_hyperparameters.optimize()
best_params_hp
```

Оптимизация лишь слегка улучшила качество.

```python
# Переопределим параметры
hp_best_params = {
    'criterion': 'mse',
    'learning_rate': 0.050182385170801695,
    'max_depth': 5,
    'max_features': 10,
    'min_samples_split': 0.579861552170576,
    'n_estimators': 296,
    'random_state': SEED
}
```

### Randomized GridSearchCV
Вместо брут-форс перебора будем случайно выбирать параметры и передавать их в модель

```python
from sklearn.model_selection import RandomizedSearchCV
```

```python
%%time
params_grid = {
    'model__learning_rate': [0.01, 0.05, 0.25, 0.5],
    'model__n_estimators': [50, 150, 250, 300],
    'model__criterion': ['friedman_mse', 'mse'],
    'model__min_samples_split': [0.1, 0.25, 0.5, 0.75, 1],
    'model__max_depth': [1, 3, 6, 9, 18],
    'model__max_features': ['auto', 'sqrt', 'log2'],
    'model__random_state': [SEED]
    }

random_search = RandomizedSearchCV(estimator=model_pipeline,
                                   param_distributions=params_grid,
                                   cv=StratifiedKFold(shuffle=True, random_state=SEED),
                                   n_iter=100,
                                   scoring='roc_auc',
                                   n_jobs=-1,
                                   random_state=SEED,
                                   verbose=2)

random_search.fit(x_train, y_train)
```

```python
# Лучшие параметры
best_params_random_cv = random_search.best_params_
best_params_random_cv
```

```python
# Для передачи в модель нужно избавится от лишних символов
best_params_random_cv = {}
for param in random_search.best_params_.keys():
    param_name = param.split('__')[1]
    param_value = random_search.best_params_[param]
    best_params_random_cv[param_name] = param_value
```

```python
# Качество с найденными парамметрами
model_pipeline = make_pipe(cat_bin_columns=x_train.select_dtypes('object').columns.to_list(),
                           num_columns=x_train.select_dtypes(['int64', 'float64']).columns.to_list(),
                           model=GradientBoostingClassifier(**best_params_random_cv),
                           cat_bin_imputer=SimpleImputer(strategy='constant', fill_value='unknown'),
                           cat_bin_encoder=TargetEncoder(),
                           num_imputer=SimpleImputer(strategy='mean'),
                           num_scaler=StandardScaler())

cv_results = cross_validate(model_pipeline,
                            X=x_train,
                            y=y_train,
                            cv=StratifiedKFold(shuffle=True, random_state=SEED),
                            scoring=['precision', 'recall', 'f1', 'roc_auc'],
                            fit_params=fit_params,
                            error_score='raise',
                            n_jobs=-1)

cv_results = pd.DataFrame(pd.DataFrame(cv_results).iloc[:,2:].mean()).T
cv_results
```

Получили качество чуть хуже, но данный метод гараздо проще и быстрее. В качестве итоговых параметров выбираем параметры найденные `HyperOpt`
