```python
import pandas as pd
import numpy as np
import warnings
import scipy.stats as sts
import matplotlib.pyplot as plt
import seaborn as sns 

from sklearn.ensemble import RandomForestClassifier, GradientBoostingClassifier
from sklearn.linear_model import LogisticRegression, RidgeClassifier
from sklearn.preprocessing import StandardScaler

import catboost as cat_boost
from catboost import CatBoostClassifier


from sklearn.impute import KNNImputer

from category_encoders import TargetEncoder
from category_encoders.cat_boost import CatBoostEncoder

from sklearn.model_selection import StratifiedKFold, cross_val_score, cross_validate

from tqdm.notebook import tqdm as log_progress

warnings.filterwarnings('ignore')
```

```python
# Необходимые функции 

def fill_na_knn(df, n_neighbors=5, weights='uniform'):
    
    knn_imputer = KNNImputer(n_neighbors=n_neighbors, weights=weights)
    
    for column in df.columns:
        df[column] = knn_imputer.fit_transform(df[[column]])
        
# Функция для кросс-валидирования моделей
def show_model_cvs(models, cv_type, x_train, y_train, metrics, is_aggregated=True):
    
    """
    Shows models results based on certain metrics and cross-validation type
    
    model - a list of models to be evaluated (Class must be passed)
    cv_type - type of cross-validation
    metrics - a list of metrics to be evaluated
    is_aggregated - returns mean estimation of the cross-validation
    
    """
    
    res_df = pd.DataFrame()
    
    for model in models:
        cv_results = cross_validate(model, X=x_train, y=y_train, cv=cv_type, scoring=metrics)
        cv_results['Model'] = str(model).split('(')[0] # вытаскиваем название обучаемоей модели
        res_df = res_df.append(pd.DataFrame(cv_results))
        
    # Чтобы название модели было первым в итоговом DF + убираем лишние столбцы
    new_columns_order = list(res_df.columns[2:-1])
    new_columns_order.insert(0, 'Model')
    
    # Возвращет информацию либо по фолдам, либо среднюю оценку
    if is_aggregated:
        return res_df[new_columns_order].groupby(by='Model').mean()
    else:
        return res_df[new_columns_order]
    
def process_test_data(df, num_columns, cat_columns):
    
    num_features_test = df[num_columns]
    cat_features_test = df[cat_columns]
    
    fill_na_knn(num_features_test)
    cat_features_test = cat_features_test.apply(lambda x: x.fillna('unknown'))
    
    return pd.concat([num_features_test, cat_features_test], axis=1)
    
    
def to_submission(data, f_name):
    result = pd.DataFrame({'Id':range(data[:,1].shape[0]),
                           'result':data[:,1]})
    result.to_csv(f_name, index=False)
    
def make_prediction(model, x_train, y_train, x_test, f_name, return_model=False):
    model.fit(x_train, y_train)
    model_preds = model.predict_proba(x_test)
    
    to_submission(model_preds,f_name)
    
    if return_model:
        return model
    
def plot_nan_ration(df, figsize=(10,5), title='Numerical Features'):
    fig = plt.figure(figsize=figsize)
    plt.hist(df.iloc[:,:-1].isna().sum()/df.shape[0])
    plt.xlabel('NaN Fraction')
    plt.ylabel('Number of Features')
    plt.title(title)
    plt.grid();
    
def calculate_crammer_coef(feat_1, feat_2, correction=True, return_p_value=True):
    """
    feat_1 - independent variable
    feat_2 - dependent variable
    
    if return_p_value=True returns p-value and V-Crammer (tuple) else only V-Crammer value (scalar)
    
    Returns feature name and its V-Crammer Value
    
    """
    confusion_matrix = pd.crosstab(feat_1, feat_2)
    n_observations = feat_1.shape[0]
    n_rows, n_colmns = confusion_matrix.shape
    
    # Confusion Matrix должна соответствовать ряду условий перед применением метода
    ratio = (np.sum((confusion_matrix.loc[:,-1] < 5)) + np.sum((confusion_matrix.loc[:,-1] < 5)))/confusion_matrix.size
    
    if ratio <= 0.2:
        # Если нужны p-values
        if return_p_value:
            chi2_res = sts.chi2_contingency(confusion_matrix, correction=correction)
            return feat_1.name, chi2_res[1], np.sqrt(chi2_res[0]/(n_observations*(min(n_rows, n_colmns)-1)))
        else:
            chi2_val = sts.chi2_contingency(confusion_matrix, correction=correction)[0]
            return feat_1.name, np.sqrt(chi2_val/(n_observations*(min(n_rows, n_colmns)-1)))
```

```python
def make_btf_heatmap(df_corrs, title='Pairwise Features Correlation'):
    fig, ax = plt.subplots(figsize=(12, 10))
    mask = np.triu(np.ones_like(df_corrs, dtype=np.bool))
    mask = mask[1:, :-1]
    corr = df_corrs.iloc[1:,:-1].copy()
    cmap = sns.diverging_palette(0, 230, 90, 60, as_cmap=True)
    sns.heatmap(corr,
                mask=mask, 
                annot=True,
                fmt=".2f",
                linewidths=5,
                cmap=cmap,
                vmin=-1,
                vmax=1, 
                cbar_kws={"shrink": .8},
                square=True)
    
    yticks = [i.upper() for i in corr.index]
    xticks = [i.upper() for i in corr.columns]
    plt.yticks(plt.yticks()[0], labels=yticks, rotation=0)
    plt.xticks(plt.xticks()[0], labels=xticks)
    plt.title(title, loc='left', fontsize=18)
    plt.show()
```

```python
make_btf_heatmap(selected_num_features.corr())
```

Видно, что признак `Var22` очень сильно коррелируют с парой других признаков, определим парог корреляции в 0.75 и удалим такие признаки

```python
def drop_cor_features(df, thresh = 0.8):
            
    corr_matrix = df.corr()
    corr_features = set()

    for i in range(len(corr_matrix.columns)):
        for j in range(i):
            if abs(corr_matrix.iloc[i,j]) >= thresh:
                col_name = corr_matrix.columns[i]
                corr_features.add(col_name)
        
    print('Original Number of Features: ', df.shape[1])
    print('Number of Higly Correlated Features: ', len(corr_features))
    df.drop(columns=corr_features, inplace=True)
    print('New Number of Features: ', df.shape[1])
    
    return df
```

В качестве основного алгоритма оптимизации модели используем модуль `hyperopt` (https://github.com/hyperopt/hyperopt/wiki/FMin)


Модель имеет очень много гиперпараметров, поэтому выбор пал на Байесовский подход к поиску гиперпараметров модели (чтобы найти наилучшую область поиска)

Некоторые важные замечания:
- Для датасетов меньше 50к наблюдений используется `boosting_type='Ordered'`
- Для таких ФП как `YetiRank, PairLogitPairwise, QueryCrossEntropy` параметр `max_depth` должен быть не более 8, для остальных до 16
- Если используется параметр `auto_class_weights`, то нельзя использовать такие параметры как `scale_pos_weight, class_weights`
- Параметр `max_bin` на GPU для ФП `PairLogitPairwise, YetiRank` не должен превышать 32
- Параметр `random_strength` не поддерживается для ФП `YetiRank, PairLogitPairwise, QueryCrossEntropy`
- Нельзя давать одинаковые `label` для параметров
- Используя модель CatBoost лучше использовать кросс-валидацию из этого модуля
- `leaf_estimation_method=Exact` is only available for Quantile, MAE and MAPE loss functions.
- Лучше кросс-валидироваться на кросс-валидации от CatBoost

Также есть ряд гиперпараметров значения которых зависят от того какие основные гиперпараметры были проинициализированы. Такие параметры мы объявим первыми, чтобы при инициализации основных параметров, генерировались и "вспомогательные".

Также объявим "Важные парметры" модели, чтобы эффективно управлять параметрами

```python
# Важные параметры CatBoost
MAX_DEPTH = 8 # Max tree depth (affects the CV time exponentially)
LOSS_FUNC = 'Logloss' 
EVAL_METRIC = 'AUC'
CAT_FEATURES = cat_features.iloc[:,:-1].columns.to_list()
TASK_TYPE = 'CPU'
OVFDET = 'Iter' # Type of overfitting detector, [default='IncToDec']
OVDWAIT = 25 # Number of iterations which overfitting detector will wait after new best error
ALGO_TYPE = tpe.suggest # Optimizator type
N_ITERS = 50 # Number of iterations for finding the best hyperparameters
CV_N_SPLITS = 3 # Number of folds for cross-validation
CV_SCORING = 'roc_auc' # metric for cross-validation
BOOST_TYPE = 'Plain' # Ordered — better quality on small datasets, but slower than the Plain scheme.

# Вложенные параметры 
grow_policy = [
    {'grow_policy':'SymmetricTree'},  # used by default 
    {'grow_policy':'Depthwise', 'min_data_in_leaf': ho_scope.int(hp.quniform('min_data_in_leaff', 1, 50, 1))},
    {'grow_policy':'Lossguide', 
         'max_leaves': ho_scope.int(hp.quniform('max_leaves', 2, 32, 1)), # don't exceed 64 
         'min_data_in_leaf': ho_scope.int(hp.quniform('min_data_in_leaf', 1, 50, 1)) # [default=1]
    }
              ]

bootstrap_type_gpu = [
    {'bootstrap_type':'Bernoulli', 'subsample':hp.uniform('subsample', 0.1, 1)},
    {'bootstrap_type':'Poisson', 'subsample':hp.uniform('subsample', 0.1, 1)}, 
    {'bootstrap_type':'Bayesian', 'bagging_temperature':hp.loguniform('bagging_temperature', np.log(1), np.log(50))}
                 ]

bootstrap_type_cpu = [
    {'bootstrap_type': 'Bernoulli', 'subsample':hp.uniform('subsample', 0.1, 1)},
    {'bootstrap_type': 'MVS'}
                     ]


auto_class_weights = ['None', 'Balanced', 'SqrtBalanced']
leaf_estimation_method = ['Newton', 'Gradient']
#leaf_estimation_backtracking_gpu = ['Armijo', 'AnyImprovement']
```

Лучшие параметры вернуться в виде словаря, в котором некторые параметры будут ссылаться на индексы (например, `grow_policy: 0`). Чтобы затем эффективно автоматически оформить все найденные гиперпараметры в словарь, первым делом значения к которым применяется `hp.choice` оформим отдельно в виде словаря, хранящий параметры со списками

Для запуска Байесовской оптимизации необходимо следующее:

**1. Необходимо определить целефую функцию, которую будем минимизировать:**

Данная функция обязательно должна принимать один аргумент (а может и нет) в виде области поиска гиперпарметров (`space_params`) и вычислять целевую метрику модели, обычно используется кросс-валидация.

**2. Область гиперпараметров:**

Обычно мы ее задаем в виде словаря, где параметры ссылаются на области значений, используя hyperopt

**3. Алгоритм минимизации**

Существуют разные алгоритмвы, самые популярные: `TPE и Simulated Anneal`

Также определим функцию `get_clean_params_dict` чтобы привести словарь `space_params` в необходимый вид для модели.

```python
def get_clean_params_dict(space_params):
    
    if space_params['bootstrap_type']['bootstrap_type'] == 'Bayesian':
        bagging_temp = space_params['bootstrap_type'].get('bagging_temperature')
        space_params['bagging_temperature'] = bagging_temp
        
    if space_params['bootstrap_type']['bootstrap_type'] in ['Poisson', 'Bernoulli']:
        subsample_val = space_params['bootstrap_type'].get('subsample')
        space_params['subsample'] = subsample_val
        
    if space_params['grow_policy']['grow_policy'] == 'Lossguide':
        max_leaves_val = space_params['grow_policy'].get('max_leaves')
        min_data_in_leaf_val = space_params['grow_policy'].get('min_data_in_leaf')
        
        space_params['max_leaves'] = max_leaves_val
        space_params['min_data_in_leaf'] = min_data_in_leaf_val
    
    if space_params['grow_policy']['grow_policy'] == 'Depthwise':
        min_data_in_leaf_val = space_params['grow_policy'].get('min_data_in_leaf')
        space_params['min_data_in_leaf'] = min_data_in_leaf_val
        
    # Перезаписываем вложенные параметры
    space_params['bootstrap_type'] = space_params['bootstrap_type']['bootstrap_type']
    space_params['grow_policy'] = space_params['grow_policy']['grow_policy']
    
    return space_params
```

```python
# Функция которая будет принимать гиперпараметры и возвращать текущий скор на кросс-валидации
def get_cv_res_with_current_hyperparams(space_params):
    
    #======================
    # Подготовка параметров
    #======================
    
    space_params = get_clean_params_dict(space_params)
    
    #================
    # Кросс-валидация
    #================
        
    cv_results = cat_boost.cv(dtrain=train_data, 
                              params=space_params, 
                              fold_count=CV_N_SPLITS, 
                              early_stopping_rounds=30, 
                              stratified=True,
                              shuffle=True,
                              partition_random_seed=SEED,
                              plot=False,
                              verbose=False)
    
    best_auc_score = cv_results['test-AUC-mean'].iloc[-1]
    
    return {'loss': -best_auc_score, 'status': STATUS_OK}    
```

```python
# Для сохранения истории 
trials = Trials()

# Данные для модели можно подготовить и таким способом
train_data = cat_boost.Pool(x_train, y_train, cat_features=CAT_FEATURES)

# Ищем оптимальные гиперпараметры
best_params = fmin(fn=get_cv_res_with_current_hyperparams,
                   space=params,
                   algo=ALGO_TYPE,
                   max_evals=N_ITERS,
                   trials=trials,
                   rstate=np.random.RandomState(SEED))
```

```python
def process_test_data(df, num_columns, cat_columns, scaler):
    
    num_features_test = df[num_columns]
    cat_features_test = df[cat_columns]
    
    
    num_features_n_cols = num_features_test.shape[1]
    
    fill_na_knn(num_features_test)
    cat_features_test = cat_features_test.apply(lambda x: x.fillna('unknown'))
    
    df = pd.concat([num_features_test, cat_features_test], axis=1)
    
    df.iloc[:,:num_features_n_cols] = scaler.transform(df.iloc[:,:num_features_n_cols])
    
    return df
```
